```python
# import libraries

import pandas as pd
import plotly.express as px
```

```python
# load csv and check how many rows and columns are there in the csv.

df=pd.read_csv('SalesData.csv')
df.shape
```

```python
df
```

```python
# checking null values

df.isnull().sum()
# There are three columns with the null values.
```

```python
# first we delete two columns because we don't need in the analysis.

df.drop(columns=['Status','unnamed1'],inplace=True)
```

```python
# deal with the null values.

df.dropna(inplace=True)
```

```python
# again check there is any null values present in the list or not.

df.isnull().sum()
```

```python
# check information of the loaded csv.

df.info()
```

```python
# check the statistical description of age,orders and amount.

df[['Age','Orders','Amount']].describe()
```

# Plotting

## Gender

```python
px.histogram(data_frame=df,x='Gender',text_auto=True)
```

```python
gender_amount=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
gender_amount
```

```python
px.bar(data_frame=gender_amount,x='Gender',y='Amount',text_auto=True)
```

## Age

```python
px.histogram(data_frame=df,x='Age Group',color='Gender',text_auto=True)
```

```python
age_amount=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
age_amount
```

```python
px.bar(data_frame=age_amount,x='Age Group',y='Amount',text_auto=True)
```

## Marital Status

```python
px.histogram(data_frame=df,x='Marital_Status',text_auto=True,color='Gender')
```

```python
marital_amount=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
marital_amount
```

```python
px.bar(data_frame=marital_amount,x='Marital_Status',y='Amount',text_auto=True)
```
